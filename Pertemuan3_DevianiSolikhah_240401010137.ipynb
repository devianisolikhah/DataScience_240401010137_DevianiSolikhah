```python
# Membaca Data Set

import pandas as pd

df = pd.read_csv(
    r"C:\Users\devia\Downloads\Praktikum Data Science\housing_dirty.csv"
)

print("✅ Dataset berhasil dimuat")
print(df.head()) 
```

```python
# Eksplorasi Dataset

df.info()

print("\nJumlah Missing Value:")
print(df.isnull().sum())

print("\nJumlah Data Duplikat:")
print(df.duplicated().sum()) 
```

```python
# Menghapus Data Duplikat

df.drop_duplicates(inplace=True)
print("\nShape setelah hapus duplikat:", df.shape) 
```

```python
# Menangani Missing Value

print("Missing Value Sebelum:")
print(df.isnull().sum())

df = df.fillna(df.median(numeric_only=True))

print("\nMissing Value Sesudah:")
print(df.isnull().sum()) 
```

```python
# Standarisasi Data

if 'kota' in df.columns:
    df['kota'] = df['kota'].str.strip().str.title()

if 'kondisi' in df.columns:
    df['kondisi'] = df['kondisi'].str.strip().str.title()

print("Standarisasi data selesai") 
```

```python
# Menangani Outlier 

import numpy as np

kolom_numerik = df.select_dtypes(include=np.number).columns

for kolom in kolom_numerik:
    Q1 = df[kolom].quantile(0.25)
    Q3 = df[kolom].quantile(0.75)
    IQR = Q3 - Q1

    bawah = Q1 - 1.5 * IQR
    atas = Q3 + 1.5 * IQR

    df = df[(df[kolom] >= bawah) & (df[kolom] <= atas)]

print("Outlier berhasil ditangani") 
```

```python
# Validasi Hasil Cleaning

print("Jumlah Data Akhir :", len(df))

print("\nMissing Value:")
print(df.isnull().sum())

display(df.head()) 
```

```python
# Menyimpan Dataset Bersih

df.to_csv("housing_clean.csv", index=False)

print("Dataset berhasil disimpan sebagai housing_clean.csv") 
```

```python
# Akses Data Dari API

import requests
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url, timeout=10)
if response.status_code == 200:
    data = response.json()
    api_df = pd.json_normalize(data)
    print("\n5 Data Pertama Dari API:")
    print(api_df[["id", "name", "email"]].head())

else:
    print("Gagal mengakses API") 
```
